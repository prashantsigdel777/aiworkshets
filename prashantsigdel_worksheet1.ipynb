```
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

add_xy = x + y
sub_xy = x - y
mul_x = x * 3
square_x = x ** 2

dot_vw = np.dot(v, w)
dot_xv = np.dot(x, v)
dot_xy = np.dot(x, y)

concat_xy_row = np.concatenate((x, y), axis=0)
concat_vw_col = np.column_stack((v, w))

try:
    bad_concat = np.concatenate((x, v))
    error_message = "No error"
except Exception as e:
    error_message = str(e)

print(add_xy)
print(sub_xy)
print(mul_x)
print(square_x)
print(dot_vw)
print(dot_xv)
print(dot_xy)
print(concat_xy_row)
print(concat_vw_col)
print(error_message)
```

```
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

invA = np.linalg.inv(A)
identity_check = A @ invA

AB = A @ B
BA = B @ A

lhs = (A @ B).T
rhs = B.T @ A.T

print(identity_check)
print(AB)
print(BA)
print(lhs)
print(rhs)
```

```
A_sys = np.array([
    [ 2, -3, 1],
    [ 1, -1, 2],
    [ 3,  1, -1]
])

B_sys = np.array([-1, -3, 9])

X1 = np.linalg.inv(A_sys) @ B_sys
X2 = np.linalg.solve(A_sys, B_sys)

print(X1)
print(X2)
```
